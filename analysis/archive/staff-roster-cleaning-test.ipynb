```python
test = staff_roster_df.groupby(
    'clean_name'
).size().reset_index()
```

```python
def get_race(df):
    name = df.clean_name
    slices = staff_roster_df[
        staff_roster_df.clean_name == name
    ]
    race_list = slices.clean_race_ethnicity.unique()
    
    if len(race_list) == 1:
        return race_list[0]
    
    if len(race_list) > 1:
        if 'multiracial' in race_list:
            return 'multiracial'
        else:
            return 'unknown'
        
def get_gender(df):
    name = df.clean_name
    slices = staff_roster_df[
        staff_roster_df.clean_name == name
    ]
    
    gender = slices.gender.value_counts().index[0]
    
    return gender
    
```

```python
test['race'] = test.apply(get_race, axis=1)
test[test.race=='multiracial']
```

```python
test['gender'] = test.apply(get_gender, axis=1)
```
